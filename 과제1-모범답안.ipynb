## 과제 1 모범 답안

```python
# 프로그램 제목 출력
print('[자동판매기 프로그램]')
print('======================')

# 지폐 및 동전 수 입력 
m1000 = int(input('입력할 1000원짜리 지폐 개수: '))  # 입력후 int 자료형으로 변환하고 변수에 저장
m500  = int(input('입력할  500원짜리 동전 개수: '))  
m100  = int(input('입력할  100원짜리 동전 개수: '))  

# 물건 값 입력
price = int(input('구입할 물건 값 입력: '))   # 입력후 int 자료형으로 변환하고 변수에 저장
print('======================')

# 총 입력 금액 계산 후 출력
money = 1000*m1000 + 500*m500 + 100*m100  # 각 입력 지폐(동전) 수에 지폐(동전)의 금액을 곱하고 모두 더해서 총 금액 계산 
print('입력한 총 금액:', money, '원')

# 거스름돈 계산 후 출력
change = money - price
print('거스름 돈 금액:', change, '원')

# 1000원짜리 지폐 거스름돈 개수 계산 및 변수 저장
c1000 = change // 1000        # 거스름돈에 1000원이 몇 개 들어있는 지 계산
change = change - c1000*1000  # 남은 거스름돈 계산: 1000원짜리 거스름돈 액수를 빼기

# 500원짜리 동전 거스름돈 개수 계산 및 변수 저장
c500   = change // 500        # 거스름돈에 500원이 몇 개 들어있는 지 계산
change = change - c500*500    # 남은 거스름돈 계산: 500원짜리 거스름돈 액수를 빼기

# 100원짜리 동전 거스름돈 개수 계산 및 변수 저장
c100   = change // 100        # 거스름돈에 100원이 몇 개 들어있는 지 계산
change = change - c100*100    # 남은 거스름돈 계산: 100원짜리 거스름돈 액수를 빼기

# 10원짜리 동전 거스름돈 개수 계산 및 변수 저장
c10    = change // 10         # 거스름돈에 10원이 몇 개 들어있는 지 계산
change = change - c10*10      # 남은 거스름돈 계산: 10원짜리 거스름돈 액수를 빼기

# 1원짜리 동전 거스름돈 개수 계산 및 변수 저장
c1     = change // 1          # 거스름돈에 1원이 몇 개 들어있는 지 계산
change = change - c1*1        # 남은 거스름돈 계산: 1원짜리 거스름돈 액수를 빼기

# 거스름돈 지폐(동전) 개수 출력
print('물건을 받으세요. 거스름돈은 다음과 같습니다')
print('1000원:', str(c1000)+'개, 500원:', str(c500)+'개, 100원:', str(c100)+'개, 10원:', str(c10)+'개, 1원:', str(c1)+'개')
```

```python
#  맨 아래 출력을 다음과 같이 할 수 있는데 쉼표(,)로 구분하여 출력하면 한칸이 띄어지기 때문에 '개' 앞에 한칸 공백이 생깁니다.
print('1000원:', c1000, '개, 500원:', c500, '개, 100원:', c100, '개, 10원:', c10, '개, 1원:', c1, '개')
```
